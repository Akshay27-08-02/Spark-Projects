## Label Encoding the dependent variable 

```python
from sklearn import preprocessing as prep
label_encoder = prep.LabelEncoder()
df['Species'] = label_encoder.fit_transform(df['Species'])
df['Species'].unique()
```

```python
df.head()
```

```python
df.drop('Id',axis=1,inplace=True)
df.head()
```

## Declaring Independent and dependent parameters under variables

```python
x=df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y=df[['Species']]
```

## Creating and Splitting train and test data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.10,random_state=0)
```

###  Creating Decision Tree Classification model

```python
from sklearn.tree import DecisionTreeClassifier as dt
model=dt(max_depth=10,random_state=45)
model.fit(x_train,y_train)
```

## Feature Scaling 

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
print(x_train)
```

## Predicting the model 

```python
predictions=model.predict(x_test)
predictions
```

## Creating classification report, confusion matrix

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

## Predicting the accuracy of the model

```python
from sklearn import metrics
print('The accuracy of the DecisionTreeClassifier is:',metrics.accuracy_score(y_test,predictions))
```

## Plotting Decision Tree

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure(figsize=(50,40))
plot_tree(model,filled=True)
```

## Plotting Confusion matrix

```python
disp = (metrics.plot_confusion_matrix(model,x_test,y_test)) 
```
